```python
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
import re
import time
```

<h1>멜론 1개월 기준 많이 들은 음악 크롤링</h1>

```python
options = webdriver.ChromeOptions()
driver = webdriver.Chrome('./chromedriver')
driver.implicitly_wait(2)
driver.get('https://www.melon.com/')

isLogin = driver.find_element_by_class_name('btn_login')

try:
    if isLogin:
    #     PH = input("멜론 아이디 : ")
        ID = 'jq3210'

    #     PW = input("비밀번호 : ")
        PW = 'power900!'

        isLogin.click()
        driver.find_element_by_xpath('//*[@title="멜론아이디 로그인"]').click()

        driver.find_element_by_name('id').send_keys(ID)
        driver.find_element_by_name('pwd').send_keys(PW)

        driver.find_element_by_class_name('btn_login03').click()

        print('로그인 성공')
    else:
        print('로그인 실패')

    driver.find_element_by_link_text('마이뮤직').click()
    driver.find_element_by_link_text('많이 들은').click()
    driver.find_element_by_link_text('1개월').click()
    time.sleep(0.5)

    html = driver.page_source
    soup = BeautifulSoup(html,'html.parser')

    trs = soup.select('tr')
    d = {}
    lidx = []
    ltitle = []
    lartist = []
    lgenre = []
    for idx,tr in enumerate(trs[1:]):
        #장르 확인
        driver.find_element_by_xpath('//*[@id="frm"]/div/table/tbody/tr[{}]/td[3]/div/div/a[1]/span'.format(idx+1)).click()
        genre = driver.find_element_by_xpath('//*[@id="downloadfrm"]/div/div/div[2]/div[2]/dl/dd[3]').text
        
        driver.back()
        driver.find_element_by_link_text('1개월').click()
        time.sleep(0.5)
        
        title = tr.select('a')[1].text.strip()
        artist = tr.select('td')[3].text.strip()
        
        lidx.append(idx+1)        #순번
        ltitle.append(title)      #제목
        lartist.append(artist)    #가수
        lgenre.append(genre)      #장르

    d['no'] = lidx
    d['title'] = ltitle
    d['artist'] = lartist
    d['genre'] = lgenre
    
    column = ['no','title','artist','genre']
    df = pd.DataFrame(d,columns = column)
except Exception as e:
    print(e)
finally:
    driver.quit()
    print('driver.quit() 완료')
```

```python
df.set_index('no',inplace=True)
df
```

<h1>최신 곡</h1><br>
여기서 이제 곡 추천해줄 예정

```python
options = webdriver.ChromeOptions()
driver = webdriver.Chrome('./chromedriver')
driver.implicitly_wait(2)
driver.get('https://www.melon.com/')

try:
    driver.find_element_by_link_text('최신 음악').click()
    time.sleep(0.2)
    
    html = driver.page_source
    soup = BeautifulSoup(html,'html.parser')
    
    trs = soup.select('tr')
    d = {}
    
    lsong_no = []
    lsong_title = []
    lsong_artist = []
    lsong_genre = []
    lsong_good = []
    
    no = 0
    for i in range(1,5):
        for idx,tr in enumerate(trs[1:]):
            driver.find_element_by_xpath('//*[@id="frm"]/div/table/tbody/tr[{}]/td[4]/div/a'.format(idx+1)).click()

            song_genre = driver.find_element_by_xpath('//*[@id="downloadfrm"]/div/div/div[2]/div[2]/dl/dd[3]').text
            song_good = driver.find_element_by_id('d_like_count').text

            driver.back()

            song_title = tr.select('a')[2].text.strip()
            song_artist = tr.select('a')[3].text.strip()

            lsong_no.append(no+1)
            lsong_title.append(song_title)
            lsong_artist.append(song_artist)
            lsong_genre.append(song_genre)
            lsong_good.append(song_good)
            
            no += 1
            
        driver.find_element_by_xpath('//*[@id="pageObjNavgation"]/div/span/a[{}]'.format(i)).click()
        time.sleep(0.5)
    
    d['no']        = lsong_no
    d['title']     = lsong_title
    d['artist']    = lsong_artist
    d['genre']     = lsong_genre
    d['good']      = lsong_good

    column = ['no','title','artist','genre','good']
    df = pd.DataFrame(d,columns = column)
    df.set_index('no',inplace = True)

except Exception as e:
    print(e)
#     continue
finally:
    driver.quit()
    print('driver.quit() 완료')
```
