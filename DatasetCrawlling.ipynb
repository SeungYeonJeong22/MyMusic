```python
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
import re
import time
```

<h1>멜론 1개월 기준 많이 들은 음악 크롤링</h1>

```python
options = webdriver.ChromeOptions()
driver = webdriver.Chrome('./chromedriver')
driver.implicitly_wait(2)
driver.get('https://www.melon.com/')

isLogin = driver.find_element_by_class_name('btn_login')

try:
    if isLogin:
    #     PH = input("멜론 아이디 : ")
        ID = 'jq3210'

    #     PW = input("비밀번호 : ")
        PW = 'power900!'

        isLogin.click()
        driver.find_element_by_xpath('//*[@title="멜론아이디 로그인"]').click()

        driver.find_element_by_name('id').send_keys(ID)
        driver.find_element_by_name('pwd').send_keys(PW)

        driver.find_element_by_class_name('btn_login03').click()

        print('로그인 성공')
    else:
        print('로그인 실패')

    driver.find_element_by_link_text('마이뮤직').click()
    driver.find_element_by_link_text('많이 들은').click()
    driver.find_element_by_link_text('1개월').click()
    time.sleep(0.5)

    html = driver.page_source
    soup = BeautifulSoup(html,'html.parser')

    trs = soup.select('tr')
    d = {}
    lidx = []
    ltitle = []
    lartist = []
    lgenre = []
    lgood = []
    for idx,tr in enumerate(trs[1:]):
        #장르 확인
        driver.find_element_by_xpath('//*[@id="frm"]/div/table/tbody/tr[{}]/td[3]/div/div/a[1]/span'.format(idx+1)).click()
        genre = driver.find_element_by_xpath('//*[@id="downloadfrm"]/div/div/div[2]/div[2]/dl/dd[3]').text
        
        driver.back()
        driver.find_element_by_link_text('1개월').click()
        time.sleep(0.5)
        
        title = driver.find_element_by_xpath('//html/body/div/div[2]/div/div/div[3]/form/div/table/tbody/tr[{}]/td[3]/div/div/a[2]'.format(idx+1)).text.strip()
        artist = driver.find_element_by_xpath('//html/body/div/div[2]/div/div/div[3]/form/div/table/tbody/tr[{}]/td[4]/div/div/a'.format(idx+1)).text.strip()
        good = driver.find_element_by_xpath('//html/body/div/div[2]/div/div/div[3]/form/div/table/tbody/tr[{}]/td[6]/div/button/span[2]'.format(idx+1)).text.strip()
        good = int(good.replace(',',''))
        
        lidx.append(idx+1)        #순번
        ltitle.append(title)      #제목
        lartist.append(artist)    #가수
        lgenre.append(genre)      #장르
        lgood.append(good)        #좋아요

    d['no'] = lidx
    d['title'] = ltitle
    d['artist'] = lartist
    d['genre'] = lgenre
    d['good'] = lgood
    
    column = ['no','title','artist','genre','good']
    df = pd.DataFrame(d,columns = column)
except Exception as e:
    print(e)
finally:
    driver.quit()
    print('driver.quit() 완료')
```

```python
df.set_index('no',inplace=True)
df
```

```python
df.to_csv('myMusic.csv')
```
